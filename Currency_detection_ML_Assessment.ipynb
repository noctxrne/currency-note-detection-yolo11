```python
!pip install ultralytics -q

import zipfile
import os

z_path = "/content/indianCurrency-notes2.zip"
e_path = "/content/dataset"

# verify zip exists
print("Zip exists:", os.path.exists(z_path))
print("Size (MB):", round(os.path.getsize(z_path) / (1024*1024), 2))

# actually extract it
with zipfile.ZipFile(z_path, 'r') as zip_ref:
    zip_ref.extractall(e_path)
```

```python
import os

for split in ['train', 'valid', 'test']:
    imgs = len(os.listdir(f'/content/dataset/{split}/images'))
    lbls = len(os.listdir(f'/content/dataset/{split}/labels'))
    match = "✓" if imgs == lbls else "✗ MISMATCH"
    print(f"{split}: {imgs} images, {lbls} labels {match}")
```

```python
from ultralytics import YOLO

model = YOLO('yolo11s.pt')
model.train(
    data='/content/dataset/data.yaml',
    epochs=80,
    imgsz=640,
    batch=16,
    name='currency_v2_yolo11',
    patience=15,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    fliplr=0.5,
    flipud=0.3,
    degrees=15,
    translate=0.1,
    scale=0.5,
    mosaic=1.0,
)
```
